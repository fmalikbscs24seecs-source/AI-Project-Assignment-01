```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tqdm import tqdm
import random

# Define dataset and plots folder
DATASET_DIR = "dataset"
PLOTS_DIR = "plots"

# Create folders if missing
os.makedirs(DATASET_DIR, exist_ok=True)
os.makedirs(PLOTS_DIR, exist_ok=True)

print("‚úÖ Folders ready!")
print("Dataset:", os.path.abspath(DATASET_DIR))
print("Plots:", os.path.abspath(PLOTS_DIR))
```

```python
# List all .wav files in dataset
audio_files = [f for f in os.listdir(DATASET_DIR) if f.endswith(".wav")]

if not audio_files:
    raise FileNotFoundError("‚ùå No .wav files found in dataset folder!")

print(f"Found {len(audio_files)} audio files.")

# Pick a random sample file for analysis
sample_file = random.choice(audio_files)
y, sr = librosa.load(os.path.join(DATASET_DIR, sample_file), sr=None)

print(f"\nüé∂ Sample file: {sample_file}")
print(f"Sample rate: {sr} Hz")
print(f"Duration: {len(y)/sr:.2f} seconds")
```

```python
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title(f"Waveform of {sample_file}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.tight_layout()
pdf_path = os.path.join(PLOTS_DIR, "waveform_sample.pdf")
plt.savefig(pdf_path)
plt.show()
print(f"‚úÖ Saved waveform plot to {pdf_path}")
```

```python
D = np.abs(librosa.stft(y))
DB = librosa.amplitude_to_db(D, ref=np.max)

plt.figure(figsize=(10, 4))
librosa.display.specshow(DB, sr=sr, x_axis="time", y_axis="log", cmap="magma")
plt.colorbar(format="%+2.0f dB")
plt.title(f"Spectrogram of {sample_file}")
plt.tight_layout()
pdf_path = os.path.join(PLOTS_DIR, "spectrogram_sample.pdf")
plt.savefig(pdf_path)
plt.show()
print(f"‚úÖ Saved spectrogram plot to {pdf_path}")
```

```python
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
print("MFCC shape:", mfcc.shape)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis="time", sr=sr)
plt.colorbar()
plt.title(f"MFCCs of {sample_file}")
plt.tight_layout()
pdf_path = os.path.join(PLOTS_DIR, "mfcc_sample.pdf")
plt.savefig(pdf_path)
plt.show()
print(f"‚úÖ Saved MFCC plot to {pdf_path}")
```

```python
results = []

for file in tqdm(audio_files, desc="Extracting MFCCs"):
    path = os.path.join(DATASET_DIR, file)
    y, sr = librosa.load(path, sr=None)
    
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc, axis=1)
    mfcc_std = np.std(mfcc, axis=1)
    
    duration = len(y) / sr
    rms = np.mean(librosa.feature.rms(y=y))
    
    results.append({
        "filename": file,
        "duration": duration,
        "rms": rms,
        **{f"mfcc_mean_{i+1}": mfcc_mean[i] for i in range(13)},
        **{f"mfcc_std_{i+1}": mfcc_std[i] for i in range(13)}
    })

df = pd.DataFrame(results)
df.to_csv("mfcc_features.csv", index=False)
print("‚úÖ Saved extracted features to 'mfcc_features.csv'")
df.head()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df["rms"], bins=15, color="lightblue", edgecolor="black")
plt.title("Audio RMS Energy Distribution (Loudness Indicator)")
plt.xlabel("RMS Energy")
plt.ylabel("Number of Audio Files")
plt.tight_layout()
pdf_path = os.path.join(PLOTS_DIR, "rms_distribution.pdf")
plt.savefig(pdf_path)
plt.show()
print(f"‚úÖ Saved RMS loudness distribution to {pdf_path}")
```

```python
mean_mfccs = df[[col for col in df.columns if "mfcc_mean" in col]].mean().to_numpy()
std_mfccs = df[[col for col in df.columns if "mfcc_std" in col]].mean().to_numpy()

# MFCC coefficient indices
x = np.arange(1, len(mean_mfccs) + 1)

# Plot with shaded std
plt.figure(figsize=(10, 5))
plt.plot(x, mean_mfccs, label="Mean MFCCs", color="tab:blue", linewidth=2, marker="o")
plt.fill_between(x, mean_mfccs - std_mfccs, mean_mfccs + std_mfccs, color="tab:blue", alpha=0.2, label="¬±1 Std Dev")

plt.title("Average MFCC Statistics Across Dataset", fontsize=14, weight='bold')
plt.xlabel("MFCC Coefficient Index", fontsize=12)
plt.ylabel("Coefficient Value", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()

# Save as PDF
pdf_path = os.path.join(PLOTS_DIR, "mfcc_statistics.pdf")
plt.savefig(pdf_path, format="pdf")
plt.show()

print(f"‚úÖ Saved MFCC statistics plot to {pdf_path}")
```

```python
print("\nüéâ Analysis Completed Successfully!")
print(f"Total files analyzed: {len(df)}")
print("‚úÖ Plots generated and saved as PDF in:")
print(os.path.abspath(PLOTS_DIR))

for f in os.listdir(PLOTS_DIR):
    if f.endswith(".pdf"):
        print(" -", f)
```
